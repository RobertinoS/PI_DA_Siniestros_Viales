### A continuacion se procedera a realizar el proceso de ETL 

- **Estandarizacion del nombre de las columnas**: Dado que los nombres de las columnas algunos estan en minusculas, y el nombre de las mismas podria ser poco intuitivo a la hora de realizar analisis posteriormente, realizo la estandarizacion de los mismos

```python
# Coloca la primera en mayúscula
df_hecho.columns = [x.capitalize() for x in df_hecho.columns]
# Reemplaza los guiones por espacios
df_hecho.columns = df_hecho.columns.str.replace('_', ' ')
# Renombra algunas columnas
df_hecho = df_hecho.rename(columns={'N victimas': 'Cantidad de victimas',
                                                      'Aaaa':'Año',
                                                      'Mm':'Mes',
                                                      'Dd':'Dia',
                                                      'Hh':'Hora en entero',
                                                      'Xy (caba)':'XY (CABA)',
                                                      'Victima': 'Victima'})
```

Verifico los cambios anteriores

```python
df_hecho.columns
```

Verifico la existencia de valores nulos y el tipo de dato 

```python
herramientas.verifica_tipo_y_nulos(df_hecho)
```

  Se observa un total de 747 valores nulos en nuestro dataframe. El 81.47% de los valores de la columna 'Altura' son nulos, seguido del 24.57% de la columna 'Cruce', 1.15% de la columna 'Dirección normalizada' y 0.14% de la columna 'Calle'.

- **Columna: 'Altura'**

  Se considerara a la columna 'Altura' como dato no relevante debido a que para que se complete la misma, el 'hecho' debe ocurrir en algun punto de una calle y segun lo anterior , la mayoria de los hechos ocurren en las esquinas.

  Adicionalmente se puede observar que la informacion de la columna 'Altura', se encuentra de forma similar en las columnas 'Lugar del hecho'y 'Dirección. A continuacion dejo constancia del mismo.

```python
df_hecho[['Lugar del hecho', 'Dirección normalizada','Altura']].head()
```

Procedo a eliminar la columna 'Altura'

```python
df_hecho=df_hecho.drop('Altura', axis=1)
df_hecho.columns
```

- **Columna: 'Cruce'**

Se observa que dicha columna, es categorica, la cual contiene 24.57% valores faltantes. La misma solo tiene datos cuando el hecho no ocurre en un cruce. Se relacionara esta columna junto con las de 'Lugar del hecho' y 'Tipo de calle', para observar los valores nulos de cruce en relación con 'Lugar del hecho' y 'Tipo de calle'

```python
df_hecho[df_hecho['Cruce'].isnull()][['Lugar del hecho', 'Tipo de calle','Cruce']].head(10)
```

Se observan los valores NO nulos de cruce en relación con 'Lugar del hecho' y 'Tipo de calle'

```python
df_hecho[df_hecho['Cruce'].notnull()][['Lugar del hecho', 'Tipo de calle','Cruce']].head(10)
```

Se observa que los datos faltantes efectivamente son los que se refieren a un hecho en algún punto de la calle que no es una esquina o intersección de dos calles. Por lo tanto, se decide modificar esta columna en un valor 'si' para los casos que son hechos en cruces de calles y 'no' en el caso que no sea un cruce de calles

```python
df_hecho['Cruce'] = np.where(df_hecho['Cruce'].notnull(), 'Si', 'No')
# Se revia la columna
df_hecho[['Lugar del hecho', 'Tipo de calle','Cruce']]
```

- **Columna: 'Datos Normalizados**

Dicha columna posee solo 8 datos nulos , por lo cual se decide modificar los valores NaN , SD (Sin Datos) debido a que no tenemos informacion de como se completo la misma. 

```python
df_hecho['Dirección normalizada'].fillna("SD", inplace=True)
df_hecho[df_hecho['Dirección normalizada']== 'SD'][['Lugar del hecho', 'Tipo de calle','Dirección normalizada']]
```

- **Columna: 'Hora'**

Convierto en datetime todos los valores de la columna 'Hora'

```python
# Se cambia el tipo de dato
df_hecho['Hora'] = df_hecho['Hora'].apply(lambda x: herramientas.convertir_a_time(x))
```

Verifico la cantidad de valores por tipo de dato en la columna 'hora'

```python
print('Tipos de datos:')
print(df_hecho['Hora'].apply(type).value_counts())
print('Registro con NoneType:')
df_hecho[df_hecho['Hora'].isna()]
```

Calculo la hora mas comun y reemplazo el unico valor None , con la misma 

```python
hora_comun = df_hecho['Hora'].mode().iloc[0]
print(f'Hora mas comun: {hora_comun}')
# Se reemplaza el valor None por la hora más común
df_hecho['Hora'].fillna(hora_comun, inplace=True)
# se verifica el tipo de dato para la columna
df_hecho['Hora'].apply(type).value_counts()
```

- **Columna: 'Hora en entero'**

```python
# Se imputa la hora moda al dato faltante
df_hecho['Hora en entero'] = df_hecho['Hora en entero'].apply(lambda x: int(hora_comun.hour) if x == "SD" else x)
# Se verifica el tipo de dato
df_hecho['Hora en entero'].apply(type).value_counts()
```

```python
# Se verifica el registro completado
df_hecho[df_hecho['Id']=='2019-0103']
```

- **Columna: 'Calle'**

```python
# Se verifica el tipo de dato
print('Tipos de datos:')
print(df_hecho['Hora en entero'].apply(type).value_counts())
# Se observa el registro con tipo de dato float
print('Registro con tipo float:')
df_hecho[df_hecho['Calle'].apply(lambda x: isinstance(x, float))]
```

```python
# Se imputa SD al dato faltante
df_hecho['Calle'].fillna('SD', inplace=True)
# Se verifica el registro completado
df_hecho[df_hecho['Id']=='2016-0151']
```

Para 'Acusado' no se observan características distintas a las esperadas. En cambio, para las categorías de 'Víctima' se observan dos categorías Objeto fijo y Peatón moto que no están en el diccionario de datos. Se observan estos registros para corregirlos.

```python
df_hecho[df_hecho['Victima'].isin(['OBJETO FIJO', 'PEATON_MOTO'])]
```

Se observan dos registros, uno por cada término distinto al del diccionario de datos. Se decide cambiarlo por OTRO dado que no hay otro dato que pueda indicar el vehículo de la victima.

```python
# Se cambia por OTRO
df_hecho['Victima'] = df_hecho['Victima'].replace({'OBJETO FIJO':'OTRO', 'PEATON_MOTO':'OTRO'})
# Se verifican esos cambios
df_hecho[df_hecho['Id'].isin(['2017-0108', '2020-0063'])]
```

- **Columnas: 'XY_(CABA)', 'Pos_x' y 'Pos_y'.**

```python
df_hecho['XY (CABA)']
```

Se observan los valores faltantes en las columnas a analaizar 

```python
df_hecho[df_hecho['XY (CABA)']=='Point (. .)']
```

Se observan 14 registros que no presentan las coordenadas planas del hecho y también se ven 12 registros que tampoco presentan los datos 'Pos x' y 'Pos y'. En este caso , dichos valores se reemplazaran por el valor **0** de modo tal que sirva para excluir el valor en el dashboard a realizar.

```python
# Se reemplazan los valores faltantes por 0
df_hecho['Pos x'] = df_hecho['Pos x'].replace('.', 0)
df_hecho['Pos y'] = df_hecho['Pos y'].replace('.', 0)
df_hecho['XY (CABA)'] = df_hecho['XY (CABA)'].replace('Point (. .)', 0)
```
